```python
import pandas as pd

df = pd.read_csv("/content/Assignment_Test.csv")
df.head()
```

```python
df['State'] = df['HDB BRANCH NAME'] + ', ' + df['HDB BRANCH STATE']
df = df.drop(['HDB BRANCH NAME', 'HDB BRANCH STATE'], axis=1)
df.head()
```

```python
df = df.dropna(subset=['State'])
df.head()
```

```python
import difflib

def compare_names(row):
  """Compares 'PAN NAME' and the combined full name."""
  # Convert potential float values to strings
  full_name = f"{str(row['FIRST NAME'])}  {str(row['LAST NAME'])}".strip()
  pan_name = str(row['Pan Name'])  # Convert 'Pan Name' to string
  similarity_ratio = difflib.SequenceMatcher(None, pan_name, full_name).ratio()
  return 'Similar' if similarity_ratio >= 0.70 else 'Not Similar'  # Adjust threshold as needed

df['Name Similarity'] =  df.apply(compare_names, axis = 1)
df.head()
```

```python
df['Cibil Score'] = pd.to_numeric(df['Cibil Score'], errors='coerce').fillna(0).astype(int)
print(df['Cibil Score'].dtype)
```

```python
bins = [300, 600, 750, 900]
labels = ['Low', 'Medium', 'High']

# Assign scores to bins for high_approval_df, use the correct column name 'Cibil Score'
df['CIBIL Score Category'] = pd.cut(df['Cibil Score'], bins=bins, labels=labels, right=False)

# Include 'Unknown' in the categories
df['CIBIL Score Category'] = df['CIBIL Score Category'].cat.add_categories('Unknown')

# Handle missing values
df['CIBIL Score Category'].fillna('Unknown', inplace=True)
df.head()
```

```python
columns_to_keep = ['UID','DEALER ID', 'Cibil Score', 'MOBILE VERIFICATION', 'TOTAL ASSET COST', 'APPLIED AMOUNT', 'EMPLOYER TYPE', 'Phone Social Premium.housing', 'Phone Social Premium.isWABusiness',  'phone_digitalage', 'phone_nameMatchScore', 'phone_phoneFootprintStrengthOverall', 'Name Similarity', 'CIBIL Score Category', 'State']
```

```python
df = df[columns_to_keep]
df.head()
```

```python
df['Phone Social Premium.housing'] = df['Phone Social Premium.housing'].fillna(0)
df['Phone Social Premium.isWABusiness'] = df['Phone Social Premium.isWABusiness'].fillna(0)
```

```python
df = df.dropna(subset=['EMPLOYER TYPE'])
df = df.drop('DEALER ID', axis=1)
df.head()
```

```python
df = df.dropna(subset=['State'])
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
def apply_label_encoder(df, columns_to_encode):
  le = LabelEncoder()
  for col in columns_to_encode:
    if col in df.columns:
      df[col] = le.fit_transform(df[col])
  return df

def scale_numerical_features(df, columns_to_scale):
  scaler = MinMaxScaler()
  df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])
  return df
```

```python
categorical_cols = ['MOBILE VERIFICATION', 'EMPLOYER TYPE', 'Application Status', 'Name Similarity', 'CIBIL Score Category', 'Phone Social Premium.housing', 'Phone Social Premium.isWABusiness', 'phone_phoneFootprintStrengthOverall', 'State'] # Moved 'State Approvals' to categorical_cols
numerical_cols = ['Cibil Score', 'TOTAL ASSET COST', 'APPLIED AMOUNT', 'phone_digitalage', 'phone_nameMatchScore']

for col in numerical_cols:
  df = scale_numerical_features(df, [col])

for col in categorical_cols:
  df = apply_label_encoder(df, [col])
```
